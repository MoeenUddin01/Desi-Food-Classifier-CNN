```python
import os
from torchvision.utils import make_grid
from torchvision.io import read_image
import torchvision.transforms.functional as F
from torchvision.transforms import Resize
```

```python
os.chdir('../')
```

```python
raw_data_dir = 'data/raw/'
```

```python
images = {}
```

```python
categories = os.listdir(raw_data_dir)
print(f'Total Categories: {len(categories)}')
```

```python
for category in categories:
    image = os.listdir(os.path.join(raw_data_dir, category))
    category_images = [os.path.join(raw_data_dir, category, img)  for img in image ]
    images[category] = category_images
```

```python
images[categories[1]][0]
```

```python
from matplotlib import pyplot as plt
import numpy as np


def show(imgs):
    if not isinstance(imgs, list):
        imgs = [imgs]
    fix, axs = plt.subplots(ncols=len(imgs), squeeze=False)
    for i, img in enumerate(imgs):
        img = img.detach()
        img = F.to_pil_image(img)
        axs[0, i].imshow(np.asarray(img))
        axs[0, i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
```

```python
resize_transform = Resize((128, 128))
```

```python
for category in categories:
   print(category)
   select_images = np.random.choice(images[category], 4)
   img_tensors = [resize_transform(read_image(img_path)) for img_path in select_images]
   img_grid = make_grid(img_tensors, nrow=5)
   show(img_grid)
```
